# -------------------------------- what the JuPyter? ---------------------------------------

## what a JuPyter Notebook does ...

- Supports e.g. Python, R, ... 
- Runs the code in cells rather than in complete scripts
- Let's you interact with and explore data directly in the Notebook 
- Let's you add additional text/information/pictures using Markdown (also mathematical notation with $\LaTeX$
- Creates reports with inline graphics and analysis, comments, etc. and export them to PDF, HTML, ... 
- Runs locally on your laptop and is displayed in your browser
- Remote access e.g. on the Google Cloud (-> DataLab) and pull data from e.g. BigQuery directly
- Allows easy collaboration on Notebooks 
- Transformation of huge datasets only limited by the machine it's running on

## ... & what a JuPyter Notebook _doesn't_

- insert formulas at the click of a button (-> some basic coding knowledge is required)
- build the report diretly into the data tables (as in Excel)
- play music
